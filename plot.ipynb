```python
import pickle
import numpy as np
import seaborn as sns
```

```python
path = "./run/exp1/exp_cifar10_fc1_0.1_sgd_epochs_50_distributed_fractional_alpha1_0_alpha2_1/training_results.pkl"
```

```python

with open('./run/exp1/exp_cifar10_fc1_0.1_sgd_epochs_50_integer_alpha1_0_alpha2_1//training_results.pkl', 'rb') as f:
    files = pickle.load(f)

loss_int = []
total_int = []
for e in files.keys():
    loss_int.append(np.mean(files[e]['train_loss']))
    total_int.extend(files[e]['train_loss'])
```

```python
with open('./run/exp1/exp_cifar10_fc1_0.1_sgd_epochs_50_distributed_fractional_alpha1_0_alpha2_1/training_results.pkl', 'rb') as f:
    files = pickle.load(f)

loss_dist = []
total_dist = []
for e in files.keys():
    loss_dist.append(np.mean(files[e]['train_loss']))
    total_dist.extend(files[e]['train_loss'])
```
