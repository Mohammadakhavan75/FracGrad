```python
from sklearn.decomposition import PCA
from dataloader import load_mnist
import numpy as np
train_loader, val_loader = load_mnist('D:/Datasets/data')

pca = PCA(10)

x_train = []
x_test = []
y_x_train = []
y_x_test = []

for X, Y in train_loader:
    for t,y in zip(X,Y):
        x_train.extend(t.permute(1,2,0).numpy().reshape(-1, 28*28))
        y_x_train.append(y)
for X, Y in val_loader:
    for t,y in zip(X,Y):
        x_test.extend(t.permute(1,2,0).numpy().reshape(-1, 28*28))
        y_x_test.append(y)

x_train = np.array(x_train)/255.
x_test = np.array(x_test)/255.
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
```

```python
np.save("train_mnist_pca10.npy", np.array(x_train)) 
np.save("test_mnist_pca10.npy", np.array(x_test)) 
np.save("train_mnist_pca10_labels.npy", np.array(y_x_train)) 
np.save("test_mnist_pca10_labels.npy", np.array(y_x_test)) 
```
